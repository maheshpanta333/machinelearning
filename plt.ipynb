# Just testing how to use subplots for logistic regression

```python

import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.exp(-x)
y4=x**2
print(x)
```

### To create subplots we use plt.subplots and pass in(no of rows in plots, no of cols in plots) and axes and fig is returned where axes is the array of individual plot

```python
fig,axes=plt.subplots(2,2,figsize=(10,8))
axes[0,0].plot(x,y1,color="b")
axes[0,0].set_title("Sine Funtion with value of x!")
axes[0,1].plot(x,y2,color="r")
axes[0,1].set_title("Cosine function with value of x!")
axes[1,0].plot(x,y3,color="g")
axes[1,0].set_title("Exponetial decay")
axes[1,1].plot(x,y3,color="m")
axes[1,1].set_title("Quadratic function!")
plt.tight_layout()
plt.show()
```
